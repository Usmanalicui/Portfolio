```python
import pandas as pd
import numpy as  np
from word2number import w2n
```

```python
df = pd.read_csv(r"E:\Study\Data Sets\dirty_ecommerce_dataset.csv")
```

```python
df.head(10)
```

```python
df_columns = list(df.columns)
print(",". join(df_columns))
```

```python
df = df.drop(['customer_name','customer_email','phone_number','comments'], axis = 1)
df.info()
```

```python
def w2n_fix(x):
    try:
        return w2n.word_to_num(x)
    except:
        return x
```

```python
df['price'] = df['price'].apply(lambda x: w2n_fix(x) if isinstance (x, str) else x)
df['quantity'] = df['quantity'].apply(lambda x: w2n_fix(x) if isinstance (x, str) else x)
df['review_rating'] = df['review_rating'].apply(lambda x: w2n_fix(x) if isinstance (x, str) else x)
```

```python
df.head(10)
```

```python
df['price'] = df['price'].astype(float)
df['quantity'] = df['quantity'].astype(float)
df['review_rating'] = df['review_rating'].astype(float)
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
#dropping dupilcates
df.drop_duplicates(inplace = True)
```

```python
#dropping duplicates of unique value columns
df.drop_duplicates("order_id", inplace = True)
```

```python
#Dropping missing values of unique columns
df.dropna(subset = ["order_id"], inplace = True )
```
