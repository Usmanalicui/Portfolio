```python
from word2number import w2n
```

```python
def w2n_fix_0(x):
    try:
        return w2n.word_to_num(x)
    except:
        return x
df['review_rating'] = df['review_rating'].apply(lambda x: w2n_fix_0(x) if isinstance(x, str) else x)
def w2n_fix_1(x):
    try:
        return w2n.word_to_num(x)
    except:
        return x
df['postal_code'] = df['postal_code'].apply(lambda x: w2n_fix_1(x) if isinstance(x, str) else x)
def w2n_fix_2(x):
    try:
        return w2n.word_to_num(x)
    except:
        return x
df['quantity'] = df['quantity'].apply(lambda x: w2n_fix_2(x) if isinstance(x, str) else x)
def w2n_fix_3(x):
    try:
        return w2n.word_to_num(x)
    except:
        return x
df['price'] = df['price'].apply(lambda x: w2n_fix_3(x) if isinstance(x, str) else x)
def w2n_fix_4(x):
    try:
        return w2n.word_to_num(x)
    except:
        return x
df['discount'] = df['discount'].apply(lambda x: w2n_fix_4(x) if isinstance(x, str) else x)
```

```python
#changing data types
df['review_rating']=df['review_rating'].astype(float)
df['price']=df['price'].astype(float)
df['quantity']=df['quantity'].astype(float)
df['discount']=df['discount'].astype(float)
df.head(10)
```

```python
df.drop_duplicates()
df.duplicated()
df.info()
```

```python
df = df.drop(['customer_email', 'phone_number','comments', 'customer_name'], axis = 1)
```

```python
df.info()
```

```python
#dropping unique ones
df = df.dropna(subset = ['order_date'])
df=df.dropna(subset = ['order_id'])
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce', dayfirst=False)
df['ship_date'] = pd.to_datetime(df['ship_date'], errors='coerce', dayfirst=False)
```

```python
df['shipping time p/d'] = (df['ship_date'] - df['order_date']).dt.days
```
